```python
# I want to import the relevant libraries
import pandas as pd
import numpy as np
```

```python
#I import my dataset to this notebook from my local directory
hdp = pd.read_csv(r'C:\Users\User\Documents\ML\HEART DX PREDICTION\heart_disease.csv')
hdp
```

```python
# Lets view the head data
hdp.head()
```

```python
#I want to clean the dataset, i dont need the column education to predict a user's heart disease status
hdp.drop(columns=['education'], inplace=True)
hdp
```

```python
#I want to remove NaNs
hdp.dropna(inplace=True)
```

```python
hdp
```

```python
#I need to convert the categorical data into numerical data using a one hot encorder
hdp['Gender'] = hdp['Gender'].replace(['Male','Female'],[1,0])
```

```python
# I check my dataset to see what i need to clean again
hdp
```

```python
#I see the prevalance of stroke is categorical
hdp.prevalentStroke.unique()
```

```python
# So I have to change it to numerical data
hdp['prevalentStroke'] = hdp['prevalentStroke'].replace(['yes','no'],[1,0])
```

```python
# I check the dataset
hdp
```

```python
hdp['Heart_ stroke'].unique()
```

```python
#I convert this heart stroke to numerical 
hdp['Heart_ stroke'] =hdp['Heart_ stroke'].replace(['yes', 'No'], [1,0])
hdp
```

```python
# I need to create an outpput dataset, i.e where the dependent variables are and the input dataset where it the independent varibles are
data_input = hdp.drop(columns=['Heart_ stroke'])
data_output = hdp['Heart_ stroke']
```

```python
# I want to count how many data are in my data output
data_output.value_counts()
```

```python
data_input
```

```python
# The dataset is very imbalance,there is oversampling 3177 and 572, so i will have to do dataset resampling,using thr imblearn
#i need to install relevant libraries for it to work
#!pip install imblearn
!pip install --upgrade scikit-learn imbalanced-learn joblib

```

```python
from imblearn.over_sampling import SMOTE
```

```python
# I set the resampling frame to 
sm = SMOTE(random_state=2)
```

```python
# i resample the data as thus
data_input, data_output = sm.fit_resample(data_input, data_output.ravel())
```

```python
# I want to count the new resampled datasets
np.unique(data_output, return_counts=True)
```

```python
# Time to train the model, I import the relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# I begin training the model, i use 80% as the trained dataset
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2)
```

```python
#Now i create the model
model = LogisticRegression()
```

```python
#Now, i fit it into the mode
model.fit(x_train, y_train)
```

```python
#I want to perform the regression on the test datasets
model.score(x_test, y_test)
```

```python
#Let me perform some prediction
input = np.array([[0,45,1,20,0,0,0,0,313,100,71,21.68,79,78]])  # I want to copy a random set of rows to run and see the prediction
```

```python
# I want to get the prediction by parsing it into a variable x
x = model.predict(input)
```
